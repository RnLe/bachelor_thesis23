```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.normal(5, 1, 10000)
x = np.random.normal(5, 100, 10000)
```

```python
plt.plot(x, y, '.', color='black')
```

```python
bins = np.histogram2d(x, y, bins=20)[0]
plt.hist(x, bins=20)
```

```python
leadingDigits = x // 1
plt.hist(leadingDigits, bins=9, range=(1, 10))
```

```python
x = np.random.normal(8, 100, 1000)

for _ in range(10000):
    x = x * np.random.normal(8, 1, x.shape)
    x = x / 10 ** (np.floor(np.log10(x)))

leadingDigits = x // 10 ** (np.floor(np.log10(x)))

plt.hist(leadingDigits, bins=9, range=(1, 9), rwidth=0.1)
```
