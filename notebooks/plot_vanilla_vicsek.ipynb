```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotx

# Lese die CSV-Datei ein
df = pd.read_csv('output.csv')

# Gruppiere die Daten nach 'N'
grouped = df.groupby('N')

plt.style.use(matplotx.styles.dracula)

# Erstelle einen Plot für jede Gruppe
for name, group in grouped:
    plt.plot(group['noise'], group['average_va'], 'x', label=f'N={name}')

# Füge eine Legende hinzu
plt.legend()

# Beschrifte die Achsen
plt.xlabel(r'Noise $\eta$')
plt.ylabel(r'Average $v_A$')

# Zeige den Plot an
plt.savefig("vicsek.png", dpi=300)
plt.show()
```

```python
# Lese die CSV-Datei ein
df = pd.read_csv('output2.csv')


plt.style.use(matplotx.styles.dracula)

# Erstelle einen Plot für jede Gruppe
plt.plot(df['N'] / df['L'] ** 2, df['average_va'], 'x')

plt.ylim(0)

# Beschrifte die Achsen
plt.xlabel(r'$\rho$')
plt.ylabel(r'Average $v_A$')

# Zeige den Plot an
plt.savefig("vicsek2.png", dpi=300)
plt.show()
```
